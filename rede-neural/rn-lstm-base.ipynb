```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import matplotlib.pyplot as plt

# 1. Importar o dataset do CSV
df = pd.read_csv('base_unificada_com_casos.csv')

# 2. Separar os dados por país
paises = df['Local'].unique()  # Obter lista de países

# 3. Função para criar os conjuntos de dados de treinamento e teste por país
def create_dataset(series, n_steps):
    X, y = [], []
    for i in range(len(series) - n_steps):
        X.append(series[i:i + n_steps])
        y.append(series[i + n_steps])
    return np.array(X), np.array(y)

# Configurações
n_steps = 5  # Número de passos de tempo para olhar para trás
future_steps = 10  # Número de anos para prever

# Dicionário para armazenar as previsões futuras por país
previsoes_futuras = {}

# 4. Iterar por cada país para treinar o modelo separadamente
for pais in paises:
    df_pais = df[df['Local'] == pais]
    
    # Selecionar as colunas relevantes para as features (excluindo 'Local' e 'Ano')
    features = df_pais[['BCG', 'Hib3', 'DTP3', 'Saneamento Seguro (%)', 'Numero Total de Imigrantes',
                        'Gini (Indice de Desigualdade no Pais)', 'Populacao', 'Taxa de Extrema Pobreza (%)',
                        'Numero de Usuario de Internet', 'Numero de Casos de Infeccoes Respiratorios e Tuberculose']].values

    # Normalizar os dados
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(features)

    # Criar os conjuntos de dados para treinamento
    X, y = create_dataset(scaled_data, n_steps)
    
    # Dividir em conjunto de treinamento e teste
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Reshape para o LSTM [amostras, passos de tempo, características]
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], X_train.shape[2]))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], X_test.shape[2]))

    # 5. Modelo LSTM
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(n_steps, X_train.shape[2])))
    model.add(Dense(X_train.shape[2]))  # Prevendo todas as features
    model.compile(optimizer='adam', loss='mse')

    # Treinando o modelo
    model.fit(X_train, y_train, epochs=200, verbose=0, validation_split=0.1)

    # Fazendo previsões
    y_pred = model.predict(X_test)
    y_pred = scaler.inverse_transform(y_pred)
    y_test = scaler.inverse_transform(y_test)

    # Calculando o erro
    mse = mean_squared_error(y_test, y_pred)
    print(f'País: {pais}, Mean Squared Error: {mse}')

    # Previsão futura
    last_sequence = scaled_data[-n_steps:]  # Pegando a última sequência para previsão
    predictions = []

    for _ in range(future_steps):  # Prever os próximos anos
        last_sequence = last_sequence.reshape((1, n_steps, X_train.shape[2]))  # Ajustando o formato para o modelo
        next_value = model.predict(last_sequence)  # Prevendo o próximo valor
        predictions.append(next_value[0])
        # Atualizando a sequência removendo o primeiro valor e adicionando o próximo valor previsto
        last_sequence = np.concatenate((last_sequence[:, 1:, :], np.expand_dims(next_value, axis=1)), axis=1)

    # Convertendo as previsões para o formato original
    predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, X_train.shape[2]))

    # Armazenando as previsões futuras
    previsoes_futuras[pais] = predictions

    # Plotando os resultados para cada país
    plt.figure(figsize=(14, 8))
    plt.plot(df_pais['Ano'], df_pais['Numero de Casos de Infeccoes Respiratorios e Tuberculose'], label='Dados Reais')
    future_years = list(range(2021, 2021 + future_steps))
    plt.plot(future_years, predictions[:, -1], label='Previsões Futuras', color='red')  # Usando a última coluna (infecções)
    plt.xlabel('Ano')
    plt.ylabel('Numero de Casos de Infeccoes Respiratorios e Tuberculose')
    plt.title(f'Previsão para {pais}')
    plt.legend()
    plt.show()

# Exibindo as previsões futuras
for pais, previsao in previsoes_futuras.items():
    future_years = list(range(2021, 2021 + future_steps))
    print(f'País: {pais}')
    print("Anos Futuros:", future_years)
    print("Previsões Futuras (Número de Casos de Infecções):", previsao[:, -1])
```
